```python
import numpy as np
from UploadData import *
from Classyfiers import *
import time
import matplotlib.pyplot as plt
from IPython.display import clear_output
from IPython import display
%matplotlib inline
```

```python
set=40

datasets, labels, train = facedataset(set)

X_mean = np.array(train).mean(axis=0)
X_std = np.array(train).std(axis=0)
train_norm = (train - X_mean)/X_std
datasets_norm = (datasets - X_mean)/X_std
```

```python
label = 4
image = 8
for label in range(32, 38):
    for image in range(6,7):
            
        #print('RP')
        fig = plt.figure(figsize=(10, 10))
        img = datasets_norm[label][image]
        failImg = datasets_norm[label-6][image]
        failImg2 = datasets_norm[label-12][image]
        failImg3 = datasets_norm[label-2][image]
        failImg4 = datasets_norm[label-4][image]
        failImg5 = datasets_norm[label-14][image-1]

        dist = [RP_clsf(img, img_c) for img_c in train_norm]
        dist = dist / np.sum(dist)


        fig.add_subplot(3, 3, 4)
        if dist.argmin() == label:
                plt.imshow(img, cmap="summer")
        else:
                plt.imshow(failImg3, cmap="autumn")
        plt.title('RP')


        img = datasets[label][image]

        fig.add_subplot(3, 3, 2)
        plt.imshow(img, cmap="gray")

        fig.add_subplot(3, 3, 1)
        plt.imshow(train[label], cmap="gray")
        plt.title('Train')


        dist = [H_clsf(img, img_c) for img_c in train]
        dist = dist / np.sum(dist)

        fig.add_subplot(3, 3, 5)
        if dist.argmin()==label:
                plt.imshow(img, cmap="summer")
        else:
                plt.imshow(failImg2, cmap="autumn")
        plt.title('H')

        dist = [G_clsf(img, img_c) for img_c in train]
        dist = dist / np.sum(dist)

        fig.add_subplot(3, 3, 6)
        if dist.argmin()==label:
                plt.imshow(img, cmap="summer")
        else:
                plt.imshow(failImg, cmap="autumn")
        plt.title('G')

        dist = [DFT_clsf(img, img_c) for img_c in train]
        dist = dist / np.sum(dist)

        fig.add_subplot(3, 3, 7)
        if dist.argmin()==label:
                plt.imshow(img, cmap="summer")
        else:
                plt.imshow(failImg4, cmap="autumn")
        plt.title('DFT')


        dist = [DCT_clsf(img, img_c) for img_c in train]
        dist = dist / np.sum(dist)

        fig.add_subplot(3, 3, 8)
        if dist.argmin()==label:
                plt.imshow(img, cmap="summer")
        else:
                plt.imshow(failImg, cmap="autumn")
        plt.title('DCT')


        dist = [Scale_clsf(img, img_c) for img_c in train]
        dist = dist / np.sum(dist)

        fig.add_subplot(3, 3, 9)
        if dist.argmin()==label:
                plt.imshow(img, cmap="summer")
        else:
                plt.imshow(failImg5, cmap="autumn")
        plt.title('Scale')

        plt.show()
        
        time.sleep(2)
        clear_output(wait=True)
```
